**(2) Considérer l’url ’http://www.python.org’,**

**(3) Récupérer le document HTML associé à l’url en utilisant urllib.request.urlopen(url).read(),**

```python
import urllib.request

url = 'http://www.python.org'
html = urllib.request.urlopen(url).read()
```

**(4) Créer un objet BeautifulSoup, représentant la structure du document HTML. Utiliser afficher l’objet créé. Il est possible d’utiliser la fonction prettify() pour obtenirun affichage amélioré.**

```python
from bs4 import BeautifulSoup

# Créez un objet BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Affichez la structure du document HTML (améliorée)
print(soup.prettify())
```

**(5) Afficher le tag title du document HTML en utilisant soup.title et accéder au texte du tag en utilisant soup.title.string**

```python
# Affichez le tag <title>
print("Tag <title>:", soup.title)

# Accédez au texte du tag <title>
print("Texte du tag <title>:", soup.title.string)
```

**(6) Afficher les liens contenu dans le document HTML (valeurs des href dans les tags a).**

```python
# Trouvez toutes les balises <a>
links = soup.find_all('a')

# Parcourez les balises <a> et affichez les valeurs des attributs href
for link in links:
    href = link.get('href')
    print("Lien trouvé:", href)
```
